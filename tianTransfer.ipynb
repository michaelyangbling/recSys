```python
import pandas as pd
import numpy as np

songs=pd.read_csv("msdchallenge/kaggle_songs.txt", sep=" ", header=None)
users=pd.read_csv("msdchallenge/kaggle_users.txt", header=None)
#song_to_track=pd.read_csv("msdchallenge/taste_profile_song_to_tracks.txt",sep="\t",header=None)
u_s_triplets=pd.read_csv("msdchallenge/kaggle_visible_evaluation_triplets.txt", sep="\t", header=None)




```

```python
songs.columns=["song_id","song_index"]
users.columns=["user_id"]
u_s_triplets.columns=["user_id","song_id","counts"]
u_s_triplets.head()
```

```python
u_s_triplets.counts.describe()
```

```python
u_s_triplets['rating'] = np.log(1+u_s_triplets['counts']) 
# Counts is highly skewed. So rating need to be scaled.
# log(1+p) eliminates 0 rating
```

```python
from sklearn.model_selection import train_test_split
subset_train,subset_test = train_test_split(u_s_triplets, train_size=0.0001, random_state=322)
```

```python
subset_train.shape
```

```python
subset_train.user_id.nunique()
```

```python
songValue=subset_train.song_id.unique()
userValue=subset_train.user_id.unique()
```

```python
a=[0.0]*142
b=[0.0]*144
for i in range(144):
    b[i]=a
```

```python
R = pd.DataFrame(b)
R.columns=songValue
R.index=userValue
```

```python
R.head()
```

```python
for i in range(len(subset_train)):
    r=subset_train.iloc[i]['user_id']
    c=subset_train.iloc[i]['song_id']
    R.loc[r][c]=subset_train.iloc[i]['rating']
    
```

```python
R
```

```python
P = np.diag(R.sum(axis=1))
Q = np.diag(R.sum(axis=0))

P_=np.diag(1/np.sqrt(R.sum(axis=1)))
Q_=np.diag(1/np.sqrt(R.sum(axis=0)))
```

```python
Gamma_U=np.matmul(np.matmul(np.matmul(np.matmul(P_,R),R.T),P_),R)
```

```python
Gamma_I=np.matmul(np.matmul(np.matmul(np.matmul(R,Q_),R.T),R),Q_)
```

```python
year1 = pd.read_csv("year1_valid_triplets_visible.txt",header=None,sep="\t")
```

```python
year1.columns=["user_id","song_id","target"]
year1['log_target']=np.log(np.add(1,year1['target']))
```

```python
year1 = year1.drop("target",axis=1)
year1 = year1.head(1000)
```

```python
user={}
song={}
for i in range(len(year1)):
    if year1.iloc[i]['user_id'] not in user:
        user[year1.iloc[i]['user_id']]=i
        year1.at[i,'user_id'] = i
    else:
        year1.at[i,'user_id'] = user[year1.iloc[i]['user_id']]
    if year1.iloc[i]['song_id'] not in song:
        user[year1.iloc[i]['song_id']]=i
        year1.at[i,'song_id']=i
    else:
        year1.at[i,'song_id'] = user[year1.iloc[i]['song_id']]
    
```

```python
from sklearn.model_selection import train_test_split
year1_train,year1_test = train_test_split(year1,train_size=0.7,random_state=41) 
```

```python
year1_train.to_csv("year1_train.csv",sep="\t",header=None,index=False)
year1_test.to_csv("year1_test.csv",sep="\t",header=None,index=False)
```
